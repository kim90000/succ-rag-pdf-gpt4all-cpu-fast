```python
import os
import numpy as np
import faiss
import torch
from transformers import AutoTokenizer, AutoModel
from PyPDF2 import PdfReader
from gpt4all import GPT4All

# Load GPT4All model
gpt4all_model = GPT4All("mistral-7b-openorca.Q4_0.gguf")

# Load embedding model
tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-small-en-v1.5")
embedding_model = AutoModel.from_pretrained("BAAI/bge-small-en-v1.5")

# Read PDF file
def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Split text into smaller chunks
def split_text(text, chunk_size=300, overlap=50):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

# Embed texts using BAAI/bge-small-en-v1.5
def embed_texts(texts):
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = embedding_model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1).numpy()
    return embeddings

# Create FAISS index
def create_faiss_index(embeddings):
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    return index

# Search for similar texts
def search_similar_texts(query_embedding, index, chunks, k=3):
    distances, indices = index.search(query_embedding, k)
    similar_texts = [chunks[i] for i in indices[0]]
    return similar_texts

# Generate answer using GPT4All
def generate_answer(query, context, max_tokens=100):
    # Truncate context to fit within the model's token limit
    max_context_tokens = 2048 - len(tokenizer.encode(query)) - 10  # Reserve tokens for query and padding
    context_tokens = tokenizer.encode(context, max_length=max_context_tokens, truncation=True)
    truncated_context = tokenizer.decode(context_tokens, skip_special_tokens=True)

    # Create prompt
    prompt = f"Context: {truncated_context}\n\nQuestion: {query}\n\nAnswer:"

    # Generate response
    response = gpt4all_model.generate(prompt, max_tokens=max_tokens)
    return response

# Path to PDF file
pdf_path = "/content/The_Lightning_Thief_-_Percy_Jackson_1-10.pdf"

# Read text from PDF
text = read_pdf(pdf_path)

# Split text into chunks
chunks = split_text(text, chunk_size=300, overlap=50)

# Embed chunks
embeddings = embed_texts(chunks)

# Create FAISS index
index = create_faiss_index(embeddings)

# Query
query = "Who is Percy?"

# Embed query
query_embedding = embed_texts([query])

# Search for similar texts
similar_texts = search_similar_texts(query_embedding, index, chunks, k=3)

# Generate answer
context = " ".join(similar_texts)
answer = generate_answer(query, context)

print("Answer:", answer)
```

```python
import numpy as np
import faiss
import torch
from transformers import AutoTokenizer, AutoModel
from PyPDF2 import PdfReader
from gpt4all import GPT4All
from nltk.tokenize import sent_tokenize
import nltk

# Download the required NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')

# Load GPT4All model
gpt4all_model = GPT4All("mistral-7b-openorca.Q4_0.gguf")

# Load embedding model
tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-small-en-v1.5")
embedding_model = AutoModel.from_pretrained("BAAI/bge-small-en-v1.5")

# Read PDF file
def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Split text into sentence-based chunks
def split_text(text, max_chunk_size=200):  # Reduced chunk size to increase the number of chunks
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) <= max_chunk_size:
            current_chunk += " " + sentence
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# Embed texts in batches
def embed_texts_in_batches(texts, batch_size=32):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i + batch_size]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            outputs = embedding_model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).numpy()
        embeddings.extend(batch_embeddings)
    return np.array(embeddings)

# Create FAISS index
def create_faiss_index(embeddings):
    dimension = embeddings.shape[1]
    nlist = min(40, len(embeddings))  # Ensure nlist <= number of training points
    quantizer = faiss.IndexFlatL2(dimension)
    index = faiss.IndexIVFFlat(quantizer, dimension, nlist, faiss.METRIC_L2)
    index.train(embeddings)
    index.add(embeddings)
    return index

# Search for similar texts
def search_similar_texts(query_embedding, index, chunks, k=3):
    distances, indices = index.search(query_embedding, k)
    similar_texts = [chunks[i] for i in indices[0]]
    return similar_texts

# Generate answer using GPT4All
def generate_answer(query, context, max_tokens=100):
    max_context_tokens = 2048 - len(tokenizer.encode(query)) - 10  # Reserve tokens for query and padding
    context_tokens = tokenizer.encode(context, max_length=max_context_tokens, truncation=True)
    truncated_context = tokenizer.decode(context_tokens, skip_special_tokens=True)
    prompt = f"Context: {truncated_context}\n\nQuestion: {query}\n\nAnswer:"
    response = gpt4all_model.generate(prompt, max_tokens=max_tokens)
    return response

# Path to PDF file
pdf_path = "/content/The_Lightning_Thief_-_Percy_Jackson_1-10.pdf"

# Read text from PDF
text = read_pdf(pdf_path)

# Split text into chunks
chunks = split_text(text, max_chunk_size=200)  # Reduced chunk size to increase the number of chunks

# Embed chunks in batches
embeddings = embed_texts_in_batches(chunks)

# Create FAISS index
index = create_faiss_index(embeddings)

# Query
query = "Who is Percy?"

# Embed query
query_embedding = embed_texts_in_batches([query])

# Search for similar texts
similar_texts = search_similar_texts(query_embedding, index, chunks, k=3)

# Generate answer
context = " ".join(similar_texts)
answer = generate_answer(query, context)

print("Answer:", answer)
```

```python
import numpy as np
import faiss
import torch
from transformers import AutoTokenizer, AutoModel
from datasets import load_dataset  # لتحميل الداتاسيت
from gpt4all import GPT4All
from nltk.tokenize import sent_tokenize
import nltk

# تحميل موارد NLTK المطلوبة
nltk.download('punkt')
nltk.download('punkt_tab')

# تحميل نموذج GPT4All
gpt4all_model = GPT4All("mistral-7b-openorca.Q4_0.gguf")

# تحميل نموذج التضمين
tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-small-en-v1.5")
embedding_model = AutoModel.from_pretrained("BAAI/bge-small-en-v1.5")

# تحميل الداتاسيت من Hugging Face
def load_huggingface_dataset(dataset_name, text_column="text"):
    dataset = load_dataset(dataset_name)
    texts = []
    for example in dataset['train']:  # افترضنا أن الداتاسيت يحتوي على قسم 'train'
        texts.append(example[text_column])
    return " ".join(texts)  # دمج النصوص في نص واحد

# تقسيم النص إلى أجزاء صغيرة
def split_text(text, max_chunk_size=200):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) <= max_chunk_size:
            current_chunk += " " + sentence
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# تضمين النصوص
def embed_texts_in_batches(texts, batch_size=32):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i + batch_size]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            outputs = embedding_model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).numpy()
        embeddings.extend(batch_embeddings)
    return np.array(embeddings)

# إنشاء FAISS index
def create_faiss_index(embeddings):
    dimension = embeddings.shape[1]
    nlist = min(40, len(embeddings))  # تأكد أن عدد العناقيد أقل من عدد النقاط
    quantizer = faiss.IndexFlatL2(dimension)
    index = faiss.IndexIVFFlat(quantizer, dimension, nlist, faiss.METRIC_L2)
    index.train(embeddings)
    index.add(embeddings)
    return index

# البحث عن النصوص المشابهة
def search_similar_texts(query_embedding, index, chunks, k=3):
    distances, indices = index.search(query_embedding, k)
    similar_texts = [chunks[i] for i in indices[0]]
    return similar_texts

# توليد الإجابة باستخدام GPT4All
def generate_answer(query, context, max_tokens=100):
    max_context_tokens = 2048 - len(tokenizer.encode(query)) - 10  # احتياطي للاستعلام والحشو
    context_tokens = tokenizer.encode(context, max_length=max_context_tokens, truncation=True)
    truncated_context = tokenizer.decode(context_tokens, skip_special_tokens=True)
    prompt = f"Context: {truncated_context}\n\nQuestion: {query}\n\nAnswer:"
    response = gpt4all_model.generate(prompt, max_tokens=max_tokens)
    return response

# اسم الداتاسيت من Hugging Face
dataset_name = "imdb"  # مثال: داتاسيت IMDB
text_column = "text"  # العمود النصي في الداتاسيت

# تحميل الداتاسيت
text = load_huggingface_dataset(dataset_name, text_column)

# تقسيم النص إلى أجزاء
chunks = split_text(text, max_chunk_size=200)

# تضمين الأجزاء
embeddings = embed_texts_in_batches(chunks)

# إنشاء FAISS index
index = create_faiss_index(embeddings)

# السؤال
query = "What is the main theme of the dataset?"

# تضمين السؤال
query_embedding = embed_texts_in_batches([query])

# البحث عن النصوص المشابهة
similar_texts = search_similar_texts(query_embedding, index, chunks, k=3)

# توليد الإجابة
context = " ".join(similar_texts)
answer = generate_answer(query, context)

print("Answer:", answer)
```
